```python
#Object oriented programing

class student(object): #Create a new class, student is a object 
    pass
s1 = student()
print(type(s1)) #Check the type of the object

s2 = student()
print(isinstance(s1, student)) #We instance check if s1 is an instance of student
print(isinstance(s2, student)) 

print(id(s1)) #Check the memory address of the object
print(id(s2))

s1 #Print the object and returns a memory address

```

```python
#Object oriented programing following the example above

s1.name = "Ada"

print(s1.name)
print(s1.__dict__) #Prints the attributes of the object
print(student.__dict__) #Prints the attributes of the class

try:                #Try to print an attribute that does not exist
    print(s2.name)
except AttributeError as err: #If the attribute does not exist
    print("Error:", err)
```

```python
#Define objects- shows a innefective example and what happens under the hood

class student(object): #Create a new class, student is a object 
    def __init__(self, name): #Initialize the object
        self.name = name
    def study(self):
        return f"{self.name} is studying"

print(student.study)

s1 = student("Ada") #Create an object

print(s1.name) #This is an attribute
print(s1.study) #This is a bound method
print(s1.study()) #This is an unbound method
print(student.study(s1)) #This is an unbound method, syntax suggared by the teacher
```

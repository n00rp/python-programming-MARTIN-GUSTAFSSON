#### Fråga 11:
 Följande kod är en del av ett program som hanterar en lista med bilar. Programmet ska kunna lägga till nya bilar, ta bort bilar och visa information om alla bilar i listan.

 class Bil:
     def __init__(self, märke, modell, år):
         self.märke = märke
         self.modell = modell
         self.år = år
 
     def __str__(self):
         return f"{self.märke} {self.modell} ({self.år})"
 
 class BilLista:
     def __init__(self):
         self.bilar = []
 
     def lägg_till_bil(self, märke, modell, år):
         bil = Bil(märke, modell, år)
         self.bilar.append(bil)
 
     def ta_bort_bil(self, index):
         if index < len(self.bilar):
             del self.bilar[index]
         else:
             print("Felaktig index")
 
     def visa_bilar(self):
         for i, bil in enumerate(self.bilar):
             print(f"{i+1}. {bil}")
 
 def main():
     bil_lista = BilLista()
 
     while True:
         print("1. Lägg till bil")
         print("2. Ta bort bil")
         print("3. Visa bilar")
         print("4. Avsluta")
 
         val = input("Välj ett alternativ: ")
 
         if val == "1":
             märke = input("Ange märke: ")
             modell = input("Ange modell: ")
             år = int(input("Ange år: "))
             bil_lista.lägg_till_bil(märke, modell, år)
         elif val == "2":
             index = int(input("Ange index för bilen du vill ta bort: ")) - 1
             bil_lista.ta_bort_bil(index)
         elif val == "3":
             bil_lista.visa_bilar()
         elif val == "4":
             break
         else:
             print("Felaktigt val")
 
 if __name__ == "__main__":
     main()

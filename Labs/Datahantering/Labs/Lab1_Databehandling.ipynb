```python

# A ) 

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

college = pd.read_csv("../Data/College.csv")
college
```

```python

# B )
college2 = pd.read_csv("../Data/College.csv", index_col=0)
college3 = college.rename({"Unnamed: 0": "College"}, axis = 1)
college3 = college3.set_index("College")

college = college3
college
```

```python

# C ) 

college.describe()
```

```python

# D ) 

pd.plotting.scatter_matrix(college[["Top10perc", "Apps", "Enroll"]], diagonal="kde")
plt.show()
```

```python

# E ) 

college.boxplot(column="Outstate", by="Private", figsize=(12,8))
plt.show()
```

```python

# F )

college["Elite"] = pd.cut(college["Top10perc"], [0,50,100], labels=["No", "Yes"])

print(college.value_counts("Elite"))

college.boxplot(column="Outstate", by="Elite", figsize=(12,8))
plt.show()
```

```python

# G )

plt.subplot(2, 2, 1)
plt.hist(college['Accept'], bins=25) 
plt.title("Accept")

plt.subplot(2, 2, 2)
plt.hist(college['Top10perc'], bins=50)
plt.title("Top10perc")

plt.subplot(2, 2, 3)
plt.hist(college['Top25perc'], bins=75)
plt.title("Top25perc")

plt.subplot(2, 2, 4)
plt.hist(college['Outstate'], bins=100)
plt.title("Outstate")

plt.tight_layout()
plt.show()
```

```python
# H )

fig = plt.figure(figsize=(12, 8))

private_grad_rate = college[college["Private"] == "Yes"]["Grad.Rate"].mean()
public_grad_rate = college[college["Private"] == "No"]["Grad.Rate"].mean()

# Plot 1: Scatterplot of Number of Applications and Accepted Applications
plt.subplot(2, 2, 1)
sns.scatterplot(x="Apps", y="Accept", hue="Private", data=college)
plt.title("Number of Applications and Accepted Applications for Private and Public Colleges")

# Plot 2: Percentage of Graduation for Private and Public Colleges
plt.subplot(2, 2, 2)
plt.bar(["Private", "Public"], [private_grad_rate, public_grad_rate])
plt.title("Percentage of Graduation")
plt.xlabel("College Type")
plt.ylabel("Percentage")

# Plot 3: Top 10 Colleges with Highest Graduation Rate
plt.subplot(2, 2, 3)
sns.barplot(x="Private", y="Grad.Rate", data=college.nlargest(10, "Grad.Rate"))
plt.title('Highest Graduation Rate')

# Plot 4: Top 10 Colleges with Most Applications
plt.subplot(2, 2, 4)
sns.barplot(x="Private", y="Grad.Rate", data=college.nlargest(10, "Apps"))
plt.title("Graduation Rate for Colleges with Most Applications")

plt.tight_layout()
plt.show()

college_sorted = college.sort_values(by='Apps', ascending=False)
print(college_sorted.head(10)[['Private', 'Apps', 'Grad.Rate']])  # Colleges with Most Applications
print(college_sorted.tail(10)[['Private', 'Apps', 'Grad.Rate']])
```



#### Analysis of College Data

In this project, I analyzed data from colleges to investigate the relationship between the number of applications, accepted applications, and graduation rates.


#### My goals were to:

Examine the relationship between the number of applications and accepted applications for private and public colleges.
Compare the graduation rates for private and public colleges.
Identify the colleges with the highest graduation rates.
Identify the colleges that received the most applications.


#### Results

The results show that:

There is a positive relationship between the number of applications and accepted applications for private and public colleges.
Private colleges have a higher average graduation rate than public colleges.
The colleges with the highest graduation rates are mostly private colleges.
The colleges that received the most applications are mostly public colleges.
